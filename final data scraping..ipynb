```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
ratings,name,review_count,location,co_type,co_age,emp_count=[],[],[],[],[],[],[]
final_df=pd.DataFrame()
for j in range(1,11):
    url="https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}".format(j)
    response=requests.get(url)
    webpage=response.text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_="company-content")
    
    for i in company:
        ratings.append((i.find('p',class_='rating').text.strip()))
        name.append(i.find('h2',class_="company-name bold-title-l").text.strip())
        review_count.append(i.find('a',class_='review-count').text.strip())
        try:
            location.append(i.find_all('p',class_="infoEntity")[1].text.strip())
        except:
            location.append(np.nan)
        co_type.append(i.find_all('p',class_="infoEntity")[0].text.strip())
        try:
            co_age.append(i.find_all('p',class_="infoEntity")[2].text.strip())
        except:
            co_age.append(np.nan)
        try:
            emp_count.append(i.find_all('p',class_="infoEntity")[3].text.strip())
        except:
            emp_count.append(np.nan)
d= {'Name':name,'Ratings':ratings,'Type':co_type,'Location':location,'Age':co_age,'Emp-count':emp_count}
#print(d)
df=pd.DataFrame(d)
final_df=final_df.append(df,ignore_index=True)
```
